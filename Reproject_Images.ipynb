```python
from osgeo import gdal
import os
```

```python

input_folder = r'N:\\Geodatenverzeichnis\\GEO-MV\\Geobasisdaten\\BDOM_HRO_LRO_ProsperRO\\TIF_Daten_10cm\\crs_5650\\export5650_dem_10cm\\'

output_folder = r'N:\\Geodatenverzeichnis\\GEO-MV\\Geobasisdaten\\BDOM_HRO_LRO_ProsperRO\\TIF_Daten_10cm\\crs_5650\\export5650_dem_10cm_reprojected\\'

EPSG_code = '5650'
```

```python
files = []
for file in os.listdir(input_folder):
    if file.endswith(".tif"):
        files.append(file)
print 'number of images: {}'.format(len(files))
```

```python
for i in range(len(files)):
    input_raster = gdal.Open(input_folder+files[i])
    code = 'EPSG:'+EPSG_code
    gdal.Warp(output_folder+files[i],input_raster,dstSRS=code)
    print '{} / {} images reprojected'.format(i+1,len(files))
    
print 'calculation of {} images done!'.format(len(files))
```
